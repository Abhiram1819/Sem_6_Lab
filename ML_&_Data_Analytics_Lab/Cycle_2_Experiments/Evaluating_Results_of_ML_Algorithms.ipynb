```python
def accuracy(tp,fp,tn,fn):
  return((tp+tn)/(tp+fp+tn+fn))
def precision(tp,fp,tn,fn):
  return(tp/(tp+fp))
def recall(tp,fp,tn,fn):
  return(tp/(tp+fn))
def sensitivity(tp,fp,tn,fn):
  return(tp/(tp+fn))
def specificity(tp,fp,tn,fn):
  return(tn/(tn+fp))
def missclserr(tp,fp,tn,fn):
  return((fp+fn)/(tp+fp+tn+fn))
```
