```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error 
```

```python
experience = np.array([2.4,5.0,1.5,3.8,8.7,3.6,1.2,8.1,2.5,5,1.6,1.6,2.4,3.9,5.4]) 
salary = np.array([2.1,4.7,1.7,3.6,8.7,3.2,1.0,8.0,2.4,6,1.1,1.3,2.4,3.9,4.8]) 
plt.scatter(experience, salary, color="red")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
a0 = 0 #intercept
a1 = 0 #Slop
lr= 0.0001 #Learning rate
iterations = 1000 # Number of iterations
error =[] # Error array to calculate cost for each iterations.
for itr in range(iterations):
    error_cost = 0
    cost_a0 = 0
    cost_a1 = 0
    for i in range (len (experience)):
        y_pred = a0+a1*experience[i] # predict value for given x error_cost = error_cost +(salary[i]-y_pred)**2
        error_cost = error_cost +(salary[i]-y_pred)**2
        for j in range (len (experience)):
            partial_wrt_a0 = -2 *(salary[j]-(a0+ a1*experience[j])) #partial derivative w.r.t a partial_wrt_a1 = (-2*experience[j])*(salary[j]-(ae+ a1*experience[j])) #partial deriva cost a cost_a0+ partial_wrt_a0 #calculate cost for each number and add
            partial_wrt_a1 = (-2*experience[j]) *(salary[j]-(a0+ a1*experience[j]))
            cost_a0 = cost_a0 + partial_wrt_a0
            cost_a1 = cost_a1 + partial_wrt_a1 #calculate cost for each number and add
        a0 = a0 - lr * cost_a0
        a1 = a1 - lr * cost_a1
    print(itr,a0,a1)
    error.append(error_cost)
```

```python
print(a0)
print(a1)
```

```python
plt.figure(figsize=(10,5))
plt.plot (np.arange (1, len(error)+1),error, color='red', linewidth = 5)
plt.title("Iteration vr error")
plt.ylabel("Error")
plt.xlabel("iterations")
```

```python
experience = experience.reshape(-1,1) 
model = LinearRegression ()
model.fit (experience, salary) 
salary_pred = model.predict (experience) 
MSE = mean_squared_error(salary, salary_pred)
print('slop', model.coef_)
print ("Intercept", model.intercept_) 
print("MSE", MSE)
```
